# 경사 하강법 (Gradient Descent) -> 최소값을 찾아감. 딥러닝에서 중요

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
w = np.arange(-5,5, 0.001)
loss = w ** 2   # 손실함수: f(w) = w^2

plt.plot(w, loss)
plt.grid()
plt.xlabel('w')
plt.ylabel('loss')
plt.show()
```

```python
# 최소값을 찾아가기 위한 임의의 w를 선택
w_init = 4
# 선택 된 w에서의 접선의 기울기(gradient)
grad = 2 * w_init
# 선택 된 최초의 w에서 기울기의 반대 방향으로 약간 이동
w_next = w_init - 0.1 * grad
print(w_next)
```

```python
w = np.arange(-5,5, 0.001)
loss = w ** 2   # 손실함수: f(w) = w^2
plt.plot(w, loss, color='darkgray')

# 최소값을 찾아가기 위한 임의의 위치
w_init = -4
loss_init = w_init ** 2 # 임의의 위치에서의 손실 함수 리턴 값
plt.scatter(w_init, loss_init, label = '0')

learning_rate = 0.9 # 학습률: gradient(기울기)에 곱해주는 값
for epoch in range(1, 11):  # 10번 반복
    grad = 2 * w_init   # 현재 위치에서의 gradient(기울기)를 계산
    w_next = w_init - learning_rate * grad  # w를 gradient의 반대 방향으로 이동.
    loss_next = w_next ** 2 # 이동한 새로운 위치(w_next)에서 손실 함수 f(w)의 리턴 값.
    plt.scatter(w_next, loss_next, label=f'{epoch}')
    w_init = w_next # 이동한 위치를 다음 이동에서 시작위치로 사용하기 위해서 

plt.grid()
plt.legend()
plt.xlabel('w')
plt.ylabel('loss')
plt.show()
```
